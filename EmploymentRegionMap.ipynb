```python
%matplotlib inline
from lds.data import Data
import pandas as pd


df = Data["Travel_2011"]
df = df.groupby("Region").sum()
df = df["Employment"]
df
```

```python
import folium
from lds.geodata import geo_regions

m = folium.Map(location=[53, -1], zoom_start=6)
m.choropleth(
    geo_data=geo_regions,
    data=df,
    columns=["Region",'Employment'],
    key_on='feature.properties.name',
    fill_color='BuPu',
    fill_opacity=0.9,
    line_opacity=0.2,
    legend_name='total employed 2011',
    highlight=True
)

m
```

```python

df = Data["Travel_2011"]
df = df.drop(['Home','Underground','Train','Bus','Taxi','Motorcycle','Drive','Passenger','Bike','Walk','Other'],axis = 1)
df = df.groupby("Region").sum()
df = df.apply(lambda x: x / x.sum() , axis = 1)["Employment"]
df
```

```python
m = folium.Map(location=[53, -1], zoom_start=6)
m.choropleth(
    geo_data=geo_regions,
    data=df,
    columns=["Region",'Employment'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.9,
    line_opacity=0.2,
    legend_name='Proportion Employed',
    highlight=True
)

m
```
