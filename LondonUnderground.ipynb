```python
import pandas as pd
import folium
from lds.data import Data
from lds.geodata import geo_london,geo_regions,geo_LAs


df = Data["Travel_2011"].set_index(["LA","Region"])
#rewrite the travel methods as propotions of the total
df = df.loc[:,"Home":"Other"].apply(lambda x: x / x.sum() , axis = 1)
df = df.query('Region == "London"')
df = df.reset_index("Region")
df.head()
```

```python
m = folium.Map(location=[51.5, -0.1], zoom_start=10)

m.choropleth(
    geo_data=geo_london,
    data=df["Underground"],
    columns=["LA",'Underground'],
    key_on='feature.properties.name',
    fill_color='BuPu',
    fill_opacity=0.9,
    line_opacity=0.2,
    legend_name='Underground Useage',
    highlight=True
)

m
```
